##Desafio Análise de Correlação:

"Suponha que você trabalha em uma empresa do setor imobiliário. Para essa empresa é muito importante precificar corretamente o imóvel, caso contrário, os corretores podem perder a venda. Como parte do time de dados, você é responsável por analisar os principais fatores que podem afetar o preço dos imóveis. Para isso, eles te enviaram a base de dados (desafio_01_preco_imoveis.csv). Como você poderia ajudá-los?"

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dados_desafio = pd.read_csv('desafio_01_preco_imoveis.csv')
dados_desafio.head(5)
```

###Analisando a correlação das variáveis com o preço do imóvel:

```python
corre=dados_desafio.corr()  #Calculando a correlação entre todas as variáveis
corre.style.background_gradient(cmap='coolwarm').set_precision(2) #Quano mais escura a cor, maior a correlação
```

Observando a matrix de correlação acima, vemos que as **3 variáveis mais correlacionadas com a coluna de preço (price)** são: sqft_living (tamanho do imóvel), grade (nota do imóvel) e sqft_above (tamanho da casa acima do solo, excluindo porão).

As **3 variáveis menos correlacionadas com o preço** são: zipcode (CEP do imóvel), long (longitude do imóvel) e condition (condição do imóvel, escala de 1 a 5).
 

### Visualizando as 5 variáveis mais correlacionadas:

## Histogramas:

```python
ax = sns.histplot(dados_desafio, x="sqft_living", color="firebrick")
ax.figure.set_size_inches(12, 6)
ax.set_title('Histograma do tamanho do imóvel', fontsize=18)
ax.set_xlabel('Tamanho do imóvel', fontsize=14)
ax.set_ylabel('Contagem ', fontsize=14)
ax
```

```python
ax = sns.histplot(dados_desafio, x="sqft_above", color = "darkorange")
ax.figure.set_size_inches(12, 6)
ax.set_title('Histograma do tamanho da casa acima do solo', fontsize=18)
ax.set_xlabel('Tamanho da casa acima do solo', fontsize=14)
ax.set_ylabel('Contagem ', fontsize=14)
ax
```
